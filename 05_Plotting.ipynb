## 05 - Plotting

We will be using the matplotlib.pyplot, this is a long library name so when we want to call the library we don't want to have to write out "matplotlib.pyplot" each time so instead we call it plt typically.

```python
#Importing matplotlib.pylot and naming it plt
import matplotlib.pyplot as plt

#I'm also going to import numpy as np
import numpy as np


#typically I will have all libraries that I import at the very top of the notebook
```

Matplotlib has a TON of functions which you can look at documentation for (aka google). We will be making a simple scatterplot of some data

```python
#lets make up some data first
x_values = range(10) #x will go from 0 to 9


#y_values3=np.random.randint(low=1,high=30,size=10)
#plt.plot(x_values,y_values3)

np.random.seed(123) #setting a seed so that the random numbers I generate in the next line are always the same
y_values = np.random.randint(low=1,high=30, size=10) #y will be random integer from 1 to 30
```

```python
#this is the most simple way of making a plot using plt which again is matplotlib
plt.plot(x_values,y_values)
```

```python
#notice the difference? 'bo' means 'blue dots', there tons of varying ways you can change colors and shapes of points
plt.plot(x_values,y_values, 'bo')
```

```python
plt.plot(x_values,y_values, 'bo-') #now adds the line in between
```

```python
#lets add axis labels and a title now
plt.plot(x_values,y_values, 'bo')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Here is my Title')
```

There are tons of more capabilities within matplotlib, like histograms, boxplots, plots in 3D (although there are better libraries for 3D). From here its googleing whatever you need.

```python
# Heres where plots could eventually get to!

from mpl_toolkits import mplot3d

ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

**#1**

Make a function, called **line_plot**, that takes in 2 lists, one called x and one called y and have 3 more inputs, which will be strings, called 'xlab', 'ylab','title'

This function will take these 5 inputs and make a "completed" line plot that has axis labels and a title

```python
import matplotlib.pyplot as plt
def line_plot(x,y,xlab,ylab,title):
    plt.plot(x,y,'^-')
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.title(title)
line_plot([1,2,3,4],[1,2,3,4],"x axis","y axis","first graph")
```

Subplots are a way to look at more than 1 graph at a time. Make your own data to make 2 scatterplots that you view in a 2x1 grid (one on top of the other)

```python
x4 = range(15)
y4 = [x for x in x4]
y5 = [x ** 2 for x in x4]
plt.figure()
#fig, ax = plt.subplots(nrows=1, ncols=2)
plt.subplot(1,2,1)
plt.plot(x4,y4,'b+')
plt.title("$x_4+y_4$")
plt.subplot(1,2,2)
plt.plot(x4,y5,'g^')
plt.title("$x_4+y^2$")
```
